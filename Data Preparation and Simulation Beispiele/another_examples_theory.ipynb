```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

White noise

```python
#define parameters for white noise
n = 1000  #number of data points in the series
mean = 0   #mean value of the white noise
std_deviation = 1  #standard deviation of the white noise

#create white noise using random numbers sampled from a normal distribution
white_noise = np.random.normal(mean, std_deviation, n)

#create a time array (you can adjust the time interval as needed)
time = np.arange(0, n)

#create a plot to visualise the white noise
plt.figure(figsize=(14, 5))
plt.plot(time, white_noise, label='White Noise')
plt.title('White Noise Simulation', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.legend()
plt.show()
```

Create time series with anomalies
1. Plot with point anomaly
2. Plot with group anomaly

Source: https://github.com/CyberLympha/Examples/blob/main/Разбор%20статей/TadGAN/TadGAN.ipynb

```python
xs = np.linspace(0, 1000, 10000)

#generate a synthetic time series as a combination of sine and cosine functions
ys = 0.5*np.sin(xs) + 0.5*np.cos(0.8*xs)
idxs = np.random.choice(len(ys), size=7)
max_ = np.max(ys)
ys[idxs] = max_*np.random.randint(7)
label = np.zeros(10000)
label[idxs] = 1
df = pd.DataFrame({'values': ys, 'labels': label})
#df.head()

#plot the time series
plt.figure(figsize=(20, 5))

#line plot for a subset of data
plt.plot(xs[2000:5000], ys[2000:5000], label='Time series')

#scatter plot for labeled points
x_labeled = [xs[i] for i in range(2000, 5000) if label[i] == 1]
y_labeled = [ys[i] for i in range(2000, 5000) if label[i] == 1]
plt.scatter(x_labeled, y_labeled, c='red', label='Anomalies')

plt.title('Time series with point anomalies', fontsize = 16)
plt.xlabel('Timepoint',fontsize = 14)
plt.ylabel('Value', fontsize = 14)
plt.legend(fontsize = 12)
plt.show()
```

```python
xs = np.linspace(0, 1000, 10000)
#generate a periodic signal with a frequency that changes anomalously
ys1 = np.cos(xs*0.5*(1+0.1*np.exp(-(xs-300)**2/1000)))

#create a label array 'label1' to mark anomalies (1) and normal data (0)
label1 = np.zeros(10000)
label1[300*10-500:300*10+500] = 1

#plot the time series

#plt.figure(figsize=(20,5))
#plt.scatter(xs, label1*(-2)-1, c='r')
#plt.plot(xs, ys1)
#plt.show()
plt.figure(figsize=(20, 5))

#scatter plot for labeled points
x_labeled = [xs[i] for i in range(2000, 6000) if label1[i] == 1]
y_labeled = [ys1[i] for i in range(2000, 6000) if label1[i] == 1]
plt.scatter(x_labeled, y_labeled, c='red', label='Anomalies')
#line plot for a subset of data
plt.plot(xs[2000:6000], ys1[2000:6000], label='Time series')

plt.title('Time series with group anomaly', fontsize = 16)
plt.xlabel('Timepoint', fontsize = 14)
plt.ylabel('Value', fontsize = 14)
plt.legend(fontsize = 12)

# Displaying the plot
plt.show()
```
